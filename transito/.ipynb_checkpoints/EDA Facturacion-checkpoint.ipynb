```python
fechas = data.index.get_level_values(1)
fecha_min = fechas.min()
fecha_max = fechas.max()
```

### Formateo ###

```python
data_flat = data.reset_index().set_index("numero_afiliacion")
```

#### Analizaremos las afiliaciones con base en 12 meses de historia.

```python
# Buscamos las casos con al menos 12 meses de historia.
nmeses_grupo = data_flat.groupby("numero_afiliacion")["facturacion"].count()
filtro_nmeses = np.where(nmeses_grupo >= 12)[0].tolist()
num_filtrados = nmeses_grupo.iloc[filtro_nmeses].index.tolist()
```

```python
len(num_filtrados)
```

```python
# Tomamos los 칰ltimos 12 meses de cada afiliaci칩n con al menos 12 meses y obtenemos la mediana de esos 12 meses.
minimo_facturacion_ejecutivo = 2500
facturacion_ultimo = data_flat.loc[num_filtrados].groupby("numero_afiliacion").tail(12)
mediana_facturacion = facturacion_ultimo.groupby("numero_afiliacion")["facturacion"].median().to_frame()
mediana_facturacion = mediana_facturacion[mediana_facturacion["facturacion"] >= minimo_facturacion_ejecutivo].dropna()
afiliaciones_unicas = mediana_facturacion.reset_index()["numero_afiliacion"].nunique() 
print(f"Analizaremos {afiliaciones_unicas} afiliaciones")
mediana_facturacion.head()
```

### Agrupaciones ### 

#### Agruparemos las cantidades de facturaci칩n en 4 variables discretas para aplicar un an치lisis de Pareto.

```python
valores_facturacion = facturacion_q3_baja["facturacion"]
```

```python
labels=["Muy Bajo", "Bajo", "Medio", "Alto" ,"Muy Alto"]
binned, bins = pd.cut(valores_facturacion, 5, labels=labels, retbins=True)
binned = binned.to_frame().reset_index()
```

```python
indices = [1, 2, 3, 4]
r_slices = np.take(bins, indices)
bin_vals = np.sort(np.append(bins, r_slices)).reshape(5, 2)
limits_df = pd.DataFrame(bin_vals)
limits_df.columns = ["limite_inf", "limite_sup"]
```

```python
limits_df["grupo"] = binned["facturacion"].unique().sort_values().tolist()
groups = binned.groupby("facturacion")["numero_afiliacion"].count()
limits_df["cantidad"] = groups.values
limits_df["porcentaje"] = limits_df["cantidad"] / limits_df["cantidad"].sum() * 100
limits_df.head()
```

```python
limits_df[["grupo", "limite_inf", "limite_sup"]].set_index("grupo")[["limite_inf", "limite_sup"]].plot(kind="barh", rot=0)
plt.title("Valores de Facturaci칩n L칤mites (Inferior y Superior) de Cada Grupo")
plt.xlabel("$ Facturaci칩n")
plt.ylabel("Grupo")
plt.show()
```

```python
limits_df[["grupo", "cantidad"]].set_index("grupo")["cantidad"].plot(kind="barh", rot=0)
plt.title("Cantidad de Afiliaciones en Cada Grupo")
plt.xlabel("# Afiliaciones")
plt.ylabel("Grupo")
plt.show()
```

```python
limits_df[["grupo", "porcentaje"]].set_index("grupo")["porcentaje"].plot(kind="barh", rot=0)
plt.title("Porcentaje de Afiliaciones en Cada Grupo")
plt.xlabel("% Afiliaciones")
plt.ylabel("Grupo")

plt.show()
```

```python
# Variables a graficar.
total = groups.values.sum() 
cumsum = groups.values.cumsum() 
percent = cumsum / total * 100

# Variables de configuraci칩n visual.
ind = np.arange(len(labels))
width = 0.98 
x = ind + 0.5 * width 

# Creamos los plots.
fig = plt.figure() 
ax1 = fig.add_subplot(111) 
ax2 = ax1.twinx()

# Graficamos valores.
ax1.bar(ind, groups.values, width=width)
ax2.plot(percent)

# Configuraciones visuales.
ax1.set_ylim(ymax=total)
ax2.set_ylim(0, 100)
ax2.axhline(80, color="gray")

# Agregamos elementos informativos. 
ax1.set_xticks(x)
ax1.set_xticklabels(labels, rotation=0)
plt.axvline(1.7, c="black")
plt.annotate(s="<-- Aqu칤 se encuentra ~80% de los datos", xy=(1.7, 50))

# Etiquetas.
ax1.set_title("An치lisis Pareto 80 / 20")
ax1.set_ylabel("# Afiliaciones")
ax1.set_xlabel("Grupos")
ax2.set_ylabel("Porcentajes") 

plt.show()
```

```python
umbral = limits_df[limits_df["grupo"] == "Bajo"]["limite_sup"].round().values[0]
umbral
```

```python
flat_median = data.reset_index().groupby("numero_afiliacion")["facturacion"].median()
proporcion_todos = flat_median[flat_median < umbral].shape[0] / flat_median.shape[0]
print(f"Eliminar칤amos el {proporcion_todos * 100:0.2f}% de los usuarios totales de la extracci칩n.")
```
